```python
import matplotlib.pyplot as plt
```

```python
xc = []
yc = []

def gradient(x):
    return 2*(x+3)

def grad(derivative_function, initial_x,learning_rate,iterations):
    x=initial_x
    for i in range(iterations):
        x=x-learning_rate*gradient(x)
        xc.append(x)
        yc.append(derivative_function(x))
    return x

def fun(x):
    return (x+3)**2

print("local minima is",grad(fun,2,0.1,20))
plt.plot(xc,yc,'ro-')
```
